# One-Hot Encoding of categorical variables

```python
!pip install pandas scikit-learn
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

data_mice = pd.read_csv("data_mice.csv")


data_mice["ROOF_TYPE_CAT"] = data_mice["ROOF_TYPE"]
data_mice["WATER_SUPPLY_TYPE_CAT"] = data_mice["WATER_SUPPLY_TYPE"]
data_mice["HEATING_TYPE_CAT"] = data_mice["HEATING_TYPE"]
data_mice["OUTER_WALLS_CAT"] = data_mice["OUTER_WALLS"]

roof_map = {
    0: "Unknown",
    1: "Flat_Roof",
    2: "Tar_Paper",
    3: "Fiber_Cement_Asbestos",
    4: "Concrete_Tiles",
    5: "Brick",
    6: "Metal_Sheets",
    7: "Thatched_Roof",
    10: "Fiber_Cement",
    11: "PVC",
    12: "Glass",
    90: "Other_Material"
}

water_map = {
    1: "Public_Water_Supply",
    2: "Private_Water_Supply",
    3: "Water_Extraction_Plant",
    4: "Well_Water",
    6: "Non_Public_Water_Supply",
    7: "Mixed_Water_Supply",
    9: "No_Water_Supply"
}

heating_map = {
    0: "Unknown",
    1: "District_Heating",
    2: "Central_Heating_Own_System",
    3: "Stove_Fireplace",
    5: "Heat_Pump",
    6: "Central_Heating_Two_Units",
    7: "Electric_Heating",
    8: "Gas_Radiators",
    9: "No_Heating",
    99: "Mixed"
}

walls_map = {
    0: "Unknown",
    1: "Brick_Walls",
    2: "Lightweight_Concrete",
    3: "Fiber_Cement_Plates",
    4: "Timber_Framing",
    5: "Wood_Cladding",
    6: "Concrete_Elements",
    8: "Metal_Plates",
    10: "Fiber_Cement",
    12: "Glass",
    90: "Other_Material"
}


data_mice["ROOF_TYPE_CAT"] = (
    data_mice["ROOF_TYPE_CAT"].map(roof_map)
)

data_mice["WATER_SUPPLY_TYPE_CAT"] = (
    data_mice["WATER_SUPPLY_TYPE_CAT"].map(water_map)
)

data_mice["HEATING_TYPE_CAT"] = (
    data_mice["HEATING_TYPE_CAT"].map(heating_map)
)

data_mice["OUTER_WALLS_CAT"] = (
    data_mice["OUTER_WALLS_CAT"].map(walls_map)
)


categorical_vars = [
    "ROOF_TYPE_CAT",
    "WATER_SUPPLY_TYPE_CAT",
    "HEATING_TYPE_CAT",
    "OUTER_WALLS_CAT"
]

encoder = OneHotEncoder(
    sparse_output=False
)

encoded_array = encoder.fit_transform(
    data_mice[categorical_vars]
)

encoded_df = pd.DataFrame(
    encoded_array,
    columns=encoder.get_feature_names_out(categorical_vars),
    index=data_mice.index
)


data_mice = data_mice.drop(
    columns=categorical_vars
)


data_mice_encoded = pd.concat(
    [data_mice, encoded_df],
    axis=1
)

print(data_mice_encoded.columns)
```
